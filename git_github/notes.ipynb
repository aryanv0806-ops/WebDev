# Git and Git Hub

## GIT

- It is free and open source distributed version control system (VCS) designed to track changes in computer files, primarily for software development.
- Git is a software and GitHub is  a service.
- VCS- Track files for changes.
- ### Repo
     - It's a folder basically that stores your files and helps  in tracking changes.

```python
git --version <# or we can use git -v #>
cd <# current directory #>
ls <# list #>
git status <# show any changes such as new file, deleted, modified in respective branch #>
ls -la <# for seeing hidden files #>
git config --global user.email "email@gmail.com"
git config --global user.name "name"
git config --global --list <# shows user details/author #>

git init <# one time per project , once the initialisation is done in the folder and .git folder is created we never use this again #>
```

- .git is a hidden folder to keep history  of all files and subfolders
- commit:- it is check point like in game.
- <b>Work flow of Git</b>: WORK=>ADD=>COMMIT.
- <B> Overall basic flow</b>: `working directory=> git add => Staging area => git commit => Repo => git push => github`

```python
<# After initialisation #>
git add <filename> <filename> ... || git add . <# for add all modified files to the staging area #>
git status
git commit -m "message"
git status
<# we will not push to git hub as we are going to study how git works. #>
```

- By defaul our branch will be the master branch if not specified or other branch is created.
