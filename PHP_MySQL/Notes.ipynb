# PHP (Hypertext Processor)

## Variables- 
  - A reusable container that holds data string, integer, float, boolean
  - Starts with dollar sign $variable_name
  - echo "hello {$name}"
  - \$ like this are to escape.. ykyk
  - ### Boolean- 
        - True, false..
## Arithmetic Operators
  - We also have null
  - everything else is same as other languages
## Get and Post method
  - $_GET, $_POST:
       - Special variables used to collect data from an html form .
       - Data is sent to the file in the action attribute of \<form> 
       - \<form action="some_file.php" method="get">
  - $_GET:
        - Data is appended to the url
        - NOT SECURE
        - char limit
        - Bookmark is possible w/values
        - GET requests can be cached
        - Better for a search page
  - $_POST:
        - Data is packaged inside the body of the HTTP request 
        - MORE SECURE
        - No data limit 
        - Cannot bookmark
        - GET requests are not cached
        - Better for submitting credentials
   - ### HOw to use it with html
         - 

```python
<?php
// A reusable container holding different data types
$name = "Aryan";          // String
$age = 20;                // Integer
$price = 299.99;          // Float
$is_logged_in = true;     // Boolean
$empty_value = null;      // Null

// String interpolation using {$variable_name}
echo "Hello, my name is {$name} and I am {$age} years old.<br>";

// Escaping the dollar sign so it prints as text instead of rendering the variable
echo "In PHP, variables start with a \$ sign. For example: \$name.<br>";

// Evaluating a boolean (True echoes as 1, False echoes as an empty string)
echo "Login status: " . ($is_logged_in ? "True" : "False") . "<br>";
?>

<?php
$x = 10;
$y = 3;

echo "Addition: " . ($x + $y) . "<br>";         // Output: 13
echo "Subtraction: " . ($x - $y) . "<br>";      // Output: 7
echo "Multiplication: " . ($x * $y) . "<br>";   // Output: 30
echo "Division: " . ($x / $y) . "<br>";         // Output: 3.3333333333333
echo "Modulus (Remainder): " . ($x % $y) . "<br>"; // Output: 1
echo "Exponentiation: " . ($x ** $y) . "<br>";  // Output: 1000 (10 to the power of 3)
?>
```
