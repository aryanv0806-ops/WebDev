## Automatic type conversion.

- When an operator is applied to the “wrong” type of value, JavaScript will quietly convert that value to the type it needs, using a set of rules that often aren’t what you want or expect. This is called `type coercion.`<br><b> An important use of '====' instead of '=='.</b><br>console.log(null == undefined);<br>
  // → true<br>
  console.log(null == 0);
  // → false<br>Expressions like 0 == false and "" == false are also true because
  of automatic type conversion. When you do not want any type conversions
  to happen, there are two additional operators: === and !==. The first tests
  whether a value is precisely equal to the other, and the second tests whether it
  is not precisely equal. So "" === false is false as expected.

## an important working of logical operators.

- `The || operator, for example, will return the value to its left when that can be converted to true and will return the value on its right otherwise.` This has the expected effect when the values are Boolean and does something analogous for values of other types
- The `&& `operator works similarly but the other way around. When the value to its left is something that converts to false, it returns that value, and otherwise it returns the value on its right.
- Another important property of these two operators is that the part to their right is evaluated only when necessary. In the case of true || X, no matter what X is—even if it’s a piece of program that does something terrible—the result will be true, and X is never evaluated. The same goes for false && X, which is false and will ignore X. This is called short-circuit evaluation.

```python
// ==========================================
// 1. Array Declaration, Accessing & Changing
// ==========================================
const cars = ["Saab", "Volvo", "BMW"];
const cars2 = new Array("Saab", "Volvo", "BMW"); // Using 'new' keyword

console.log(cars[0]); // Output: "Saab"

cars[0] = "Opel";     // Changing the first element
console.log(cars);    // Output: ["Opel", "Volvo", "BMW"]

// ==========================================
// 2. Length Property
// ==========================================
const fruits = ["Banana", "Orange", "Apple", "Mango"];
console.log(fruits.length); // Output: 4

fruits.length = 2;          // You can also use length to truncate an array
console.log(fruits);        // Output: ["Banana", "Orange"]

// ==========================================
// 3. Converting to Strings
// ==========================================
const fruitsStr = ["Banana", "Orange", "Apple"];

console.log(fruitsStr.toString());   // Output: "Banana,Orange,Apple"
console.log(fruitsStr.join(" _ "));  // Output: "Banana _ Orange _ Apple"

// ==========================================
// 4. Adding & Removing (End of Array)
// ==========================================
const stack = ["Banana", "Orange"];

stack.push("Apple");      // Adds "Apple" to the end
console.log(stack);       // Output: ["Banana", "Orange", "Apple"]

let poppedItem = stack.pop(); // Removes the last element
console.log(poppedItem);      // Output: "Apple"
console.log(stack);           // Output: ["Banana", "Orange"]

// ==========================================
// 5. Adding & Removing (Beginning of Array)
// ==========================================
const queue = ["Banana", "Orange"];

queue.unshift("Lemon");   // Adds "Lemon" to the start
console.log(queue);       // Output: ["Lemon", "Banana", "Orange"]

let shiftedItem = queue.shift(); // Removes the first element
console.log(shiftedItem);        // Output: "Lemon"
console.log(queue);              // Output: ["Banana", "Orange"]

// ==========================================
// 6. Concatenating (Merging)
// ==========================================
const arr1 = [1, 2];
const arr2 = [3, 4];

const merged = arr1.concat(arr2, 5, "Peter");
console.log(merged); // Output: [1, 2, 3, 4, 5, "Peter"]

// ==========================================
// 7. copyWithin 
// (Overwrites existing values without changing length)
// ==========================================
const copyFruits = ["Banana", "Orange", "Apple", "Mango", "Kiwi"];

// Copy to index 2, taking elements from index 0 to 2
copyFruits.copyWithin(2, 0, 2);
console.log(copyFruits); // Output: ["Banana", "Orange", "Banana", "Orange", "Kiwi"]

// ==========================================
// 8. Flattening Arrays
// ==========================================
const multiArr = [[1, 2], [3, 4], [5, 6]];
console.log(multiArr.flat()); // Output: [1, 2, 3, 4, 5, 6]

const nums = [1, 2, 3];
// Maps each element, then flattens the result into a new array
const flatMapped = nums.flatMap(x => [x, x * 10]);
console.log(flatMapped); // Output: [1, 10, 2, 20, 3, 30]

// ==========================================
// 9. Splice (Mutates original array)
// splice(insert_index, elements_to_remove, new_items...)
// ==========================================
const spliceArr = ["Banana", "Orange", "Apple", "Mango"];

// At index 2, remove 0 items, and insert "Lemon", "Kiwi"
let removedItems = spliceArr.splice(2, 0, "Lemon", "Kiwi");
console.log(spliceArr);    // Output: ["Banana", "Orange", "Lemon", "Kiwi", "Apple", "Mango"]
console.log(removedItems); // Output: [] (Nothing was removed)

// Removing an item using splice
spliceArr.splice(0, 1);
console.log(spliceArr);    // Output: ["Orange", "Lemon", "Kiwi", "Apple", "Mango"]

// ==========================================
// 10. Slice (Returns a NEW array)
// slice(start_index, end_index_not_included)
// ==========================================
const originalArr = ["Banana", "Orange", "Lemon", "Apple", "Mango"];

const citrus = originalArr.slice(1, 3); // Extracts index 1 and 2
console.log(citrus);       // Output: ["Orange", "Lemon"]
console.log(originalArr);  // Output: ["Banana", "Orange", "Lemon", "Apple", "Mango"] (Unchanged)
```

### Prototype, **proto** and Polyfills

## Prototype:

- ### What is Prototype?

  - It is a blueprint or master object that other objects can inherit properties and methods from, creating a "chain" of inheritance; when you look for something in an object and don't find it, JavaScript automatically checks its prototype, then its prototype's prototype, and so on, until it finds it or hits the end of the chain, allowing shared functionality without duplicating code.

- ### What is the need of prototype in JS?? :\
  - Look at this example., if i make an object like this<br>
    function Person(name) {<br> this.name = name;<br> this.sayHello = function () {<br> console.log("Hello " + this.name);<br> };<br>}<br>
    So, what's the problem here?/ 🤔🤔<br>
    - if we call Person(), each time sayHello() will be created--> MEMORY WASTAGE<br>
      So instead of this, JS has a common place where this things are stored and that place is ---> `Prototype`<br>
  - So how we use prototype? A way of using it is:<br>
    function Person(name) {<br> this.name = name;<br>Person.prototype.sayHello = function () {<br> console.log("Hello " + this.name);<br>};<br><br>
    <b>name</b> → har object ka apna<br><b>sayHello</b> → shared<br> <b>Prototype</b> = shared object for methods<br>
- ### Internally kya hota h???
  <br> let p= new Person("Aryan");

    1. Creates an empty Object { }
    2. Uska `__proto__` set karta h-->> Person.prototype
    3. Constructor call karta h.
  - p1--> **proto** → Person.prototype

- <b>Prototype Chain</b>: This prototype object has a prototype of its own, and so on until an object is reached with null as its prototype.`Null` has no prototype and acts as the final link in this chain.
